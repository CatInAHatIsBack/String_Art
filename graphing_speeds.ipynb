```python
import matplotlib.pyplot as plt
import numpy as np


with open('time_output.txt', 'r') as f:
    delays = [float(line.strip()) for line in f]

print(delays)
```

```python
# Recalculating the cumulative time for each step
cumulative_time = np.cumsum(delays)
speeds = 1 / np.array(delays)


plt.figure(figsize=(12,6))

# Plot for speed vs. cumulative time
plt.subplot(1, 2, 1)
plt.plot(cumulative_time, speeds, marker='o', color='blue')
plt.title("Stepper Motor Speed vs. Time")
plt.xlabel("Time (ms)")
plt.ylabel("Speed (reciprocal of delay)")
plt.grid(True)

# Plot for delay vs. cumulative time
plt.subplot(1, 2, 2)
plt.plot(cumulative_time, delays, marker='o', color='red')
plt.title("Stepper Motor Delay vs. Time")
plt.xlabel("Time (ms)")
plt.ylabel("Delay (ms)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Array of delays in milliseconds

# Calculating cumulative time and speeds
cumulative_time = np.cumsum(delays)
speeds = 1 / np.array(delays)

# Calculating changes in speed and time
delta_speeds = np.diff(speeds)  # Change in speed between each step
delta_time = np.diff(cumulative_time) / 1000  # Change in time between each step in seconds

# Calculating acceleration (change in speed over change in time)
acceleration = delta_speeds / delta_time

# To plot acceleration, we need to align it with the appropriate time. We'll use the midpoint between each time interval.
midpoints_time = (cumulative_time[:-1] + cumulative_time[1:]) / 2

# Generating the plot for speed, delay, and acceleration vs. cumulative time
plt.figure(figsize=(18,6))

# Plot for speed vs. cumulative time
plt.subplot(1, 3, 1)
plt.plot(cumulative_time, speeds, marker='o', color='blue')
plt.title("Stepper Motor Speed vs. Time")
plt.xlabel("Time (ms)")
plt.ylabel("Speed (reciprocal of delay)")
plt.grid(True)

# Plot for delay vs. cumulative time
plt.subplot(1, 3, 2)
plt.plot(cumulative_time, delays, marker='o', color='red')
plt.title("Stepper Motor Delay vs. Time")
plt.xlabel("Time (ms)")
plt.ylabel("Delay (ms)")
plt.grid(True)

# Plot for acceleration vs. cumulative time
plt.subplot(1, 3, 3)
plt.plot(midpoints_time, acceleration, marker='o', color='green')
plt.title("Stepper Motor Acceleration vs. Time")
plt.xlabel("Time (ms)")
plt.ylabel("Acceleration (speed change per second)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))

delays_in_seconds = np.array(delays) / 1e6  # Convert microseconds to seconds

# Calculate points per second (frequency)
points_per_second = 1 / delays_in_seconds

# Generate the cumulative time
cumulative_time_seconds = np.cumsum(delays_in_seconds)

# Generating the plot for points per second vs. cumulative time
plt.subplot(1, 2, 1)
plt.plot(cumulative_time_seconds, points_per_second, marker='o', color='blue', linewidth=1, markersize=4)
plt.title("Points Per Second vs. Cumulative Time")
plt.xlabel("Cumulative Time (seconds)")
plt.ylabel("Points Per Second")
plt.grid(True)
plt.show()

# Plot for acceleration vs. cumulative time
plt.subplot(1, 2, 2)
plt.plot(midpoints_time, acceleration, marker='o', color='green',linewidth=.1, markersize=.1)
plt.title("Stepper Motor Acceleration vs. Time")
plt.xlabel("Time (ms)")
plt.ylabel("Acceleration (speed change per second)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

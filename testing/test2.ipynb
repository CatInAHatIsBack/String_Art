```python
import numpy as np

def linear_interpolation(point1, point2, num_points):
    x_values = np.linspace(point1[0], point2[0], num_points)
    y_values = np.linspace(point1[1], point2[1], num_points)
    return list(zip(x_values.astype(int), y_values.astype(int)))
```

```python
def bresenham_line(x0, y0, x1, y1):
    points = []
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)
    x, y = x0, y0
    sx = -1 if x0 > x1 else 1
    sy = -1 if y0 > y1 else 1
    if dx > dy:
        err = dx / 2.0
        while x != x1:
            points.append((x, y))
            err -= dy
            if err < 0:
                y += sy
                err += dx
            x += sx
    else:
        err = dy / 2.0
        while y != y1:
            points.append((x, y))
            err -= dx
            if err < 0:
                x += sx
                err += dy
            y += sy      
    points.append((x, y))
    return points
```

```python
import matplotlib.pyplot as plt

# Define the start and end points
start = (1, 1)
end = (8, 5)

# Calculate the distance
d = int(np.hypot(end[0]-start[0], end[1]-start[1]))

# Get the points for each method
points_bresenham = bresenham_line(start[0], start[1], end[0], end[1])
points_interpolation = linear_interpolation(start, end, d)

# Plot the points
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Bresenham Line')
plt.plot([p[0] for p in points_bresenham], [p[1] for p in points_bresenham], 'bo')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.title('Linear Interpolation')
plt.plot([p[0] for p in points_interpolation], [p[1] for p in points_interpolation], 'ro')
plt.grid(True)

plt.show()
```
